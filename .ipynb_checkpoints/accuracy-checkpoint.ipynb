```python
import re

locations = ["daylight", "indoor-daylight", "indoor-dark"]
ESP32 = {}
current_key = None

for filename in locations:

    full_filename = filename + ".txt"
    with open(full_filename, "r") as file:
        ESP32[filename] = {}
        
        for line in file:
            if line.strip() == '':
                continue
            # Check if the line matches the regex pattern (any number followed by "cm" with or without a space)
            elif re.search(r'\d+\s?cm', line):
                ESP32[filename][line.strip()] = []
                current_key = line.strip()
                
            elif current_key != None:
                ESP32[filename][current_key].append(int(line.strip()))
                
            else:
                print("No key!")

for key in ESP32.keys():
    print(key)
    print(ESP32[key].keys())
    print()
```

```python
import statistics

ESP32_averages = {}

for location in locations:
    ESP32_averages[location] = {}
    
    for distance in ESP32[location].keys():
        ESP32_averages[location][distance] = statistics.mean(ESP32[location][distance]) / 10

for key in ESP32_averages.keys():
    print(key)
    print(ESP32_averages[key].items())
    print()
```

```python
import matplotlib.pyplot as plt
import itertools

markers = itertools.cycle(('o', 's', 'v'))
colors = itertools.cycle(("orange", "red", "black"))

# Function to convert distance strings to numerical values
def extract_distance(dist_str):
    return int(dist_str.split()[0])

distances = sorted(next(iter(ESP32.values())).keys())

# Initialize a plot
plt.figure(figsize=(10, 6))

# Plot data for each location
for location, measurements in ESP32_averages.items():
    distances = list(extract_distance(dist) for dist in measurements.keys()) # x-values
    measured_values = list(measurements.values()) # y-values
    marker = next(markers)
    color = next(colors)
    plt.plot(distances, measured_values, marker, color=color, label=location)

# Determine the overall range of distances for the reference line
all_distances = sorted(set(extract_distance(dist) for measurements in ESP32_averages.values() for dist in measurements.keys()))

min_distance = min(all_distances)
max_distance = max(all_distances)

# Plot actual distances as a reference line
plt.plot([min_distance, max_distance], [min_distance, max_distance], 'k--', label='Actual Distance')

# Add labels and title
plt.xlabel('Actual Distance (cm)')
plt.ylabel('Measured Distance (cm)')
plt.title('Measured vs Actual Distance')
plt.legend()

plt.xticks(ticks=all_distances, labels=[f'{d} cm' for d in all_distances], rotation=45)

plt.axis([0, 65, 0, 65])

# Show the plot
plt.grid(True)
plt.show()
```
